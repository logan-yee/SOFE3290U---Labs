##5. Set Up Great Expectations Context and Data Source

```python
context = gx.get_context()

data_source = context.data_sources.add_pandas("pandas_datasource")
```

##6. Define and Create a Data Batch

```python
data_asset = data_source.add_dataframe_asset(name="labels_asset")

batch_definition = data_asset.add_batch_definition_whole_dataframe("labels_batch")

batch = batch_definition.get_batch(batch_parameters={"dataframe": df})

print(batch)
```

### Expectation 1

```python
# Image filenames should match a pattern like "A_001.png"
expectation_1 = gx.expectations.ExpectColumnValuesToMatchRegex(
    column="Occluded_Image_view",
    regex=r"^A_\d{3}\.png$"
)
```

### Validate Data Against Expectation 1

```python
result_1 = batch.validate(expectation_1)
print("Expectation 1 - Image filename format (Occluded_Image_view):")
print(f"  Success: {result_1.success}")
print(f"  Unexpected count: {result_1['result'].get('unexpected_count', 'N/A')}")
print(f"  Unexpected values: {result_1['result'].get('partial_unexpected_list', [])}")
```

### Expectation 2

```python
# Pedestrian bounding box X coordinates should be positive pixel values
expectation_2 = gx.expectations.ExpectColumnValuesToBeBetween(
    column="pedestrianLocationX_TopLeft",
    min_value=0,
    max_value=1920  # assuming max image width of 1920px
)
```

### Validate Data Against Expectation 2

```python
result_2 = batch.validate(expectation_2)
print("Expectation 2 - Pedestrian X coordinate within image bounds:")
print(f"  Success: {result_2.success}")
print(f"  Unexpected count: {result_2['result'].get('unexpected_count', 'N/A')}")
print(f"  Unexpected values: {result_2['result'].get('partial_unexpected_list', [])}")
```

### Expectation 3

```python
# Critical columns should have no null/missing values
expectation_3 = gx.expectations.ExpectColumnValuesToNotBeNull(
    column="Ground_Truth_View"
)
```

### Validate Data Against Expectation 3

```python
result_3 = batch.validate(expectation_3)
print("Expectation 3 - No nulls in Ground_Truth_View column:")
print(f"  Success: {result_3.success}")
print(f"  Unexpected count: {result_3['result'].get('unexpected_count', 'N/A')}")
print(f"  Unexpected values: {result_3['result'].get('partial_unexpected_list', [])}")
```
